# Pandas Basics: NYC Airbnb Dataset

Practicing core pandas operations: loading data, indexing, filtering, 
sorting, groupby, merging, and handling missing values.

```python
# importing pandas library
import pandas as pd

# loading the dataset
df = pd.read_csv('../data/AB_NYC_2019.csv') # while reading path .. means go back one directory, similarly ../../ means go back two directories 

# displaying the first 3 rows of the dataset, default head() displays 5 rows
df.head(3)
```

```python
# displaying the shape of the dataset, shape returns a tuple of (rows, columns)
print(df.shape)
# displaying the column names of the dataset, columns returns an Index object
print(df.columns.tolist())
# info() methods provides a concise summary of the DataFrame, including the number of non-null entries, data types, and memory usage
df.info()
```

```python
# describing the dataset, describe() method generates descriptive statistics that summarize the central tendency, dispersion and shape of a dataset’s distribution, excluding NaN values
df.describe()
```

## Indexing and Selecting Data

```python
# Select a single column
# prices = df['price'] # using single square brackets returns a Series
prices = df[['price']] # using double square brackets returns a DataFrame 
prices.head()
```

```python
# Select multiple columns
subset = df[['name', 'neighbourhood', 'price', 'room_type']]
subset.head()
```

```python
# .loc: label-based selection 
# Meaning you have to specify the row and column labels to select data from a DataFrame. 
# It allows you to select rows and columns by their labels, 
# which can be useful when you want to select specific rows or columns based on their names.
df.loc[0:5, ['name', 'price']] # displays rows 0 to 5 and columns 'name' and 'price'
```

```python
# .iloc: position-based selection
# It allows you to select rows and columns by their integer positions,
# which can be useful when you want to select specific rows or columns based on their index positions.
df.iloc[0:5, 0:4] # displays rows 0 to 5 and columns 0 to 4 (excluding column 4)
```

## Filtering Data

```python
# Listings under $100
cheap_listings = df[df['price'] < 100] # returns a DataFrame with rows that satisfy the condition

# Multiple conditions
manhattan_cheap = df[(df['neighbourhood_group'] == 'Manhattan') & (df['price'] < 150)] # multiple conditions can be combined using the & or | operator, and each condition must be enclosed in parentheses

print(f"Cheap listings: {len(cheap_listings)}")
print(f"Cheap Manhattan listings: {len(manhattan_cheap)}")
```

## Sorting Data

```python
# Sort by price 
#sort() vs sort_values(): sort() is a method of the DataFrame class that sorts the DataFrame in place, while sort_values() is a method of the Series class that returns a new Series sorted by the specified values.
# df.sort_values('price').head() #default is ascending order
# for descending
df.sort_values('price', ascending=False).head()
```

```python
# Sort by multiple columns
df.sort_values(['neighbourhood_group', 'price'], ascending=[True, False]).head(7) # first sort by neighbourhood_group in ascending order, then by price in descending order
```

## GroupBy and Aggregation

```python
# Average price by neighbourhood group
# groupby method is used to group the data based on a specific column, in this case, 'neighbourhood_group'.,
# mean() method is then applied to the 'price' column of each group to calculate the average price for each neighbourhood group.
df.groupby('neighbourhood_group')['price'].mean().sort_values(ascending=False)
```

```python
# Multiple stats at once
df.groupby('neighbourhood_group')['price'].agg(['mean', 'median', 'count', 'min', 'max'])
```

```python
# Average price by neighbourhood group AND room type
df.groupby(['neighbourhood_group', 'room_type'])['price'].mean().reset_index()
```

## Handling Missing Values

```python
# Count missing values per column
df.isnull().sum().sort_values(ascending=False) # isnull method returns a DataFrame of the same shape as df, with True for missing values and False for non-missing values.
```

```python
# 'reviews_per_month' is often missing when there are no reviews yet — fill with 0
df['reviews_per_month'] = df['reviews_per_month'].fillna(0)

# 'name' and 'host_name' missing — likely just absent, drop rows missing critical info if needed
df_clean = df.dropna(subset=['name'])

print(f"Original rows: {len(df)}, after dropping missing names: {len(df_clean)}")
```

## Summary / Observations

- Manhattan and Brooklyn have the highest average listing prices among NYC boroughs; 
  Staten Island and the Bronx are noticeably cheaper.
- Entire homes/apartments command a higher average price than private or shared rooms 
  across all boroughs.
- `reviews_per_month` had a meaningful number of missing values — likely listings with 
  no reviews yet, so filling with 0 was more appropriate than dropping rows.
- `name` and `host_name` had a small number of missing values, likely just incomplete 
  listings; dropped rather than imputed since these are identifying fields.
